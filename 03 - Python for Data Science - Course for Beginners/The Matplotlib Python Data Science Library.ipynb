```python
# Matplotlib is a very powerful package to plot/visualize data
# .pyplot being the must used module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# For the sake of example, let's create 50 linear values between 0 and 10
x = np.linspace(0,10,50)
y = np.sin(x)
# and let's plot sin(x) as a function of x
plt.plot(x,y)
```

```python
# Instead of plotting a line, we can plot the same data as points
plt.scatter(x,y,)
```

```python
# You can display multiple data on the same plot, and use a bunch of available options
plt.plot(x,np.sin(x),'-r')
plt.plot(x,np.sin(x-1),'--c')
plt.plot(x,np.sin(x-2),':g')
```
